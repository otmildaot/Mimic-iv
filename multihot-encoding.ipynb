```python
import pandas as pd

# 1. read file CSV
file_path = 'C:/Users/Alan Jose/Downloads/mimic-iv-3.1/mimic-iv-3.1/hosp/preprocess/admissions_11.csv'  # ganti sesuai nama file kamu
df = pd.read_csv(file_path, low_memory=False)

# 2. Defined all multi-label column and label dictionary
multilabel_columns = {
    'cvd': {
        1: 'cvd_acute_rheumatic',
        2: 'cvd_chronic_rheumatic',
        3: 'cvd_hypertensive',
        4: 'cvd_ischemic',
        5: 'cvd_other_heart',
        6: 'cvd_cerebrovascular',
        7: 'cvd_arterial',
        8: 'cvd_other_circulatory'
    },
    'familyhistory': {
        1: 'family_stroke',
        2: 'family_cvd'
    },
}

# 3. Fungsi multi-hot encoding in general
def multi_hot_encode(cell, label_keys):
    vector = [0] * len(label_keys)
    if pd.isna(cell) or str(cell).strip() == '0':
        return vector
    for code in str(cell).split(','):
        try:
            idx = int(code.strip()) - 1
            if 0 <= idx < len(label_keys):
                vector[idx] = 1
        except ValueError:
            pass
    return vector

# 4. Terapkan ke setiap kolom multi-label
for col, label_dict in multilabel_columns.items():
    label_names = list(label_dict.values())
    df[label_names] = df[col].apply(lambda x: pd.Series(multi_hot_encode(x, label_names)))

# 5. save file in a new file
df.to_csv('C:/Users/Alan Jose/Downloads/mimic-iv-3.1/mimic-iv-3.1/hosp/preprocess/data_encoded.csv', index=False)
```
